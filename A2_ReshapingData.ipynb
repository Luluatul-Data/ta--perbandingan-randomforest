```python
#import library

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#membaca dataset

df = pd.read_csv('CR300Series_2_Daily.csv')
```

```python
#membaca 10 data pertama pada dataset

df.head(10)
```

```python
#jumlah baris dan kolom
df.shape
```

```python
#untuk mengetahui deskripsi setiap variabel
df.describe()
```

```python
#mengecek nilai null
df.isna().sum()
```

```python
#menghapus baris 1 sampai 3 karena semuanya berisi nan

df.drop(index = range(0, 3), inplace = True)
print (df)
```

```python
#merapikan index setelah di hapus nilai nan-nya

df.reset_index(drop=True,
               inplace=True)
```

```python
#melihat 20 data teratas

df.head(20)
```

```python
#membuat data baru dengan variabel yang dibutuhkan

df = df[['TIMESTAMP','BP_mbar_Avg', 'AirTC_Avg', 'AirTC_Max', 'AirTC_Min', 'RH', 'WS_ms_Avg', 'WS_ms_Max','WindDir','Rain_mm_Tot']]
df
```

```python
deskripsi = df.describe()
print(deskripsi)
```

```python
#membuat perulangan colom sebanyak 21 kali
num_columns = 21
dfs = [df.copy()]
for i in range(1, num_columns):
    shifted_df = df.shift(-i)
    shifted_df.columns = [f'{col}_{i}' for col in shifted_df.columns]
    dfs.append(shifted_df)

# Menggabungkan semua DataFrames yang terbentuk
final_df = pd.concat(dfs, axis=1)

```

```python
print(final_df)
```

```python
#menampilkan 21 kolom pada data yang terbentuk saat ini

print(final_df.iloc[:, 0:21])
```

```python
#melihat null pada data baru

final_df.isna().sum()
```

```python
#menghapus setiap null

df = final_df.dropna(axis=0)
```

```python
#menyimpan hasil transformasi ke dalam file CSV

final_df.to_csv('output.csv', index=False)
```
