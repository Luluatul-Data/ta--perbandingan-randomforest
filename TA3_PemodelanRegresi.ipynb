```python
#import library

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#membaca dataset

df = pd.read_csv('output_finish.csv')
```

```python
#menampilkan 20 data teratas pada dataset

df.head(20)
```

```python
#menampilkan 10 kolom pertama pada dataset

print(df.iloc[:, 0:10])
```

```python
#menampilkan nilai null pada dataset

df.isna().sum()
```

```python
#menampilkan variabel rain yang akan digunakan sebagai target

print (df.Rain_mm_Tot_7)
```

PEMODELAN

```python
#import library

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#membuat fitur x dan y
x = df.drop('Rain_mm_Tot_7', axis=1)
y = df['Rain_mm_Tot_7']
```

```python
#menampilkan fitur x dan y yang telah terbentuk
print ("fitur X : \n", x , "\n")
print ("fitur Y : \n", y , "\n")
```

SPLITTING

```python
#membagi data training dan testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```

```python
#menampilkan data training dan testing

print("x_train:\n", x_train, "\n")
print("x_test:\n", x_test, "\n")
print("y_train:\n", y_train, "\n")
print("y_test:\n", y_test, "\n")
```

FINE-TUNING

```python
#import library

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python

rf = RandomForestRegressor()

# Hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}

# Grid Search
grid_search = GridSearchCV(estimator=rf,
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=3,
                           n_jobs=-1)
grid_search.fit(x_train, y_train)

# Evaluasi hasil
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Tampilkan hasil
print("Best Parameters:", grid_search.best_params_)
print("RMSE:", rmse)
```

```python
#membangun model random forest regression dengan hyperparameter yang didapatkan dari fine-tuning

rf_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=4,
    max_features='log2',
    bootstrap=True,
)
```

```python
# melatih data menggunakan data latih

rf_model.fit(x_train, y_train)
```

```python
# melakukan prediksi menggunakan data uji

y_pred = rf_model.predict(x_test)
```

```python
# menampilkan hasil prediksi

print("Hasil prediksi : ", y_pred)
```

```python
# Menampilkan perbandingan antara nilai asli dan prediksi

comparison_df = pd.DataFrame({
    'Data Asli': y_test,
    'Prediksi': y_pred
})

print(comparison_df)
```

Threshold

```python

#fungsi klasifikasi berdasarkan curah hujan harian (dalam mm, bertipe float)
def klasifikasi_hujan(mm):
    if 0 <= mm <= 5.0:
        return 'Cerah'
    elif 5.0 < mm <= 20.0:
        return 'Hujan Ringan'
    elif 20.0 < mm <= 50.0:
        return 'Hujan Sedang'
    elif 50.0 < mm <= 100.0:
        return 'Hujan Lebat'
    elif mm > 100.0:
        return 'Hujan Sangat Lebat'
    else:
        return 'Di luar klasifikasi'

#menerapkan fungsi klasifikasi ke kolom DataFrame

comparison_df['Kategori Data Asli'] = comparison_df['Data Asli'].apply(klasifikasi_hujan)
comparison_df['Kategori Prediksi'] = comparison_df['Prediksi'].apply(klasifikasi_hujan)

#menampilkan 30 baris pertama hasil klasifikasi
print(comparison_df.head(30))
```

Mean Absolute Error

```python
#import library
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import mean_absolute_error, confusion_matrix, ConfusionMatrixDisplay
```

```python
#  menghitung akurasi model menggunakan MAE

mae = mean_absolute_error(y_test, y_pred)
print(f"MAE : {mae:.2f}")
```

Confusion Matrix

```python
#mengambil label aktual dan prediksi dari dataframe
y_true = comparison_df['Kategori Data Asli']
y_pred = comparison_df['Kategori Prediksi']

#menentukan urutan label kategori sesuai standar BMKG
labels = ["Cerah", "Hujan Ringan", "Hujan Sedang", "Hujan Lebat","Hujan Sangat Lebat"]

#menghitung prediksi masing-masing kategori pada confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=labels)

# Tampilkan confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap="Blues", xticks_rotation=45)
plt.title("Confusion Matrix Klasifikasi Cuaca (Standar BMKG)")
plt.grid(False)
plt.tight_layout()
plt.savefig("grafik31.png")
plt.show()
```

```python
# Hitung akurasi
accuracy = accuracy_score(y_true, y_pred)*100

# Hitung F1 Score (macro dan weighted)
f1_macro = f1_score(y_true, y_pred, labels=labels, average='macro')*100
f1_weighted = f1_score(y_true, y_pred, labels=labels, average='weighted')*100

# Tampilkan hasil
print(f"Akurasi: {accuracy:.2f}%")
print(f"F1 Score (Macro): {f1_macro:.2f}%")
print(f"F1 Score (Weighted): {f1_weighted:.2f}%")
```
